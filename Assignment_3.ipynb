```python
import numpy as np
from scipy.interpolate import lagrange
x = np.array([-5., -3., -1.,  1.,  3.,  5.])
y = (1/(1 + x**2))
poly = lagrange(x, y)



from numpy.polynomial.polynomial import Polynomial
Polynomial(poly.coef[::-1]).coef

import matplotlib.pyplot as plt
def function(x):
   return (1/(1 + x**2))
x_new = np.arange(-5, 5.1, 0.1)
plt.scatter(x, y, label='data')
plt.plot(x_new, Polynomial(poly.coef[::-1])(x_new), label='Polynomial')
plt.plot(x_new, function(x_new), color='red')

plt.show()

plt.show()
```

maximum error is at x = 0, and looks to be  a litte more than  .4

```python
import numpy as np
import sys
#here the interval is[a,b] so a must be < b
def bisection_method_function(f,a,b,tol):
  if np.sign(f(a)) == np.sign(f(b)):
    raise Exception ("this can't work because the end points have the same sign")

  m = (a+b)/2
  print(f(m))

  if np.abs(f(m)) < tol:
    print ("%r is close enough to our root" %(m))

  elif f(m) < 0 :
    return bisection_method_function(f,m,b,tol)
  elif f(m) > 0:
    return bisection_method_function(f,a,m,tol)


g = lambda x : x**3 +x**2 +2

bisection_method_function(g,-2,3,0.0000001)


```

```python
def Newton_method(x,g,dg,tol):#have to do it like in the book where derivative is found first by person.otherwise annoying error.if i had more time I would do it so that python would find derivative.
  x_old = x
  if np.abs(g(x_old)) < tol:
    return x_old
  else:
    x_new = x_old - (g(x_old)/dg(x_old))
    print(x_new) #outputting iterations
    return Newton_method(x_new,g,dg,tol)
```

```python
j = lambda x : x**3 +x**2 +2
dj = lambda x : 3*x**2 + 2*x

Newton_method(1,j,dj,.1)

```

```python
def Secant_method(x,y,g,tol):#have to do it like in the book where derivative is found first by person.otherwise annoying error.if i had more time I would do it so that python would find derivative.
  x_old = x
  y_old = y
  if np.abs(g(x_old)) < tol:
    return x_old
  else:
    x_new = x_old - (g(x_old)*(x_old-y_old))/(g(x_old)-g(y_old))
    print(x_new) #outputting iterations
    return Secant_method(x_new,x_old,g,tol)
```
