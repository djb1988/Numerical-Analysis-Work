```python
import math
import random, cmath
import numpy
N = 10000
incircle = 0
count = 0
other_array = []#empty array to test
x_array= [] #empty array to put x values in
y_array = []
for i in range(1, 10001):

  other_array.append(i)
  value = (1/i)
  x_array.append(value)
print(x_array)
while (count<N):
  x = random.uniform(0, 1)
  y = random.uniform(0, 1)
  if math.sqrt((x)**2 +(y)**2) <= 1 :
    incircle = incircle+1
  count = count+1
  estimate = (incircle/10000) *4

  error = (math.pi - estimate)
  y_array.append(error)
print(y_array)
print(other_array)


print( "Pi is estimated as {} ".format(estimate))





```
