```python
import numpy as np
import matplotlib.pyplot as plt


h = .1
x = np.linspace(0, 1 + h, 12)
   # Get A
A = np.zeros((12, 12))
A[0, 0] = 1
A[11, 11] = 1
A[11, 10] = 0
A[ 11 , 9] = -1
for i in range(1, 11):
    A[i, i-1] = 1
    A[i, i] = -2- h**2
    A[i, i+1] = 1

    # Get b
b = np.zeros(12)
b[0] = 1



empty_array = []#using this to store vector of real solutions info

for i in range(0,12):
  v = ((np.exp(2-x[i]) + np.exp (x[i])))/(1 + np.exp(2))
  empty_array.append(v)


results = np.linalg.solve(A, b)

plt.plot(x,results, color ='r', label = 'approximation')





```

Compare this graph of the appoximation above to the graph of the real solution below.very similar.


```python
plt.plot(x,empty_array, color='g',label='real solution')
```
