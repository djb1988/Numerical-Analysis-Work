```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

# Create a cube
fig = plt.figure()
ax = fig.add_subplot(121, projection='3d')
r = [-1, 1]
X, Y = np.meshgrid(r, r)
ax.plot_surface(X, Y, 1, alpha=0.5, color='red')
ax.plot_surface(X, Y, -1, alpha=0.5, color='red')
ax.plot_surface(X, 1, Y, alpha=0.5, color='red')
ax.plot_surface(X, -1, Y, alpha=0.5, color='red')
ax.plot_surface(1, X, Y, alpha=0.5, color='red')
ax.plot_surface(-1, X, Y, alpha=0.5, color='red')
ax.set_title('Cube')

# Create a sphere
ax2 = fig.add_subplot(122, projection='3d')
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
x = np.outer(np.cos(u), np.sin(v))
y = np.outer(np.sin(u), np.sin(v))
z = np.outer(np.ones(np.size(u)), np.cos(v))
ax2.plot_surface(x, y, z, color='blue', alpha=0.5)
ax2.set_title('Sphere')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate points on the surface of the cube
points = []
for x in [-1, 1]:
    for y in np.linspace(-1, 1, 20):
        for z in np.linspace(-1, 1, 20):
            points.append([x, y, z])
            points.append([y, x, z])
            points.append([y, z, x])
            points.append([z, y, x])
points = np.array(points)

# Function to map points from the cube to the sphere
def map_to_sphere(point):
    norm = np.linalg.norm(point)
    return point / norm

# Function to map points from the sphere back to the cube
def map_to_cube(point):
    # Here, the point on the sphere is already normalized
    norm = np.linalg.norm(point)
    return point * norm  # This is a conceptual step to illustrate the mapping

# Map the points to the sphere and back to the cube
sphere_points = np.array([map_to_sphere(p) for p in points])
cube_points = np.array([map_to_cube(p) for p in sphere_points])

# Plot the original cube points
fig = plt.figure(figsize=(18, 6))
ax1 = fig.add_subplot(131, projection='3d')
ax1.scatter(points[:, 0], points[:, 1], points[:, 2], c='r', marker='o')
ax1.set_title('Unit Cube')
ax1.set_xlim([-1.5, 1.5])
ax1.set_ylim([-1.5, 1.5])
ax1.set_zlim([-1.5, 1.5])

# Plot the sphere points
ax2 = fig.add_subplot(132, projection='3d')
ax2.scatter(sphere_points[:, 0], sphere_points[:, 1], sphere_points[:, 2], c='b', marker='o')
ax2.set_title('Unit Sphere')
ax2.set_xlim([-1.5, 1.5])
ax2.set_ylim([-1.5, 1.5])
ax2.set_zlim([-1.5, 1.5])

# Plot the points mapped back to the cube
ax3 = fig.add_subplot(133, projection='3d')
ax3.scatter(cube_points[:, 0], cube_points[:, 1], cube_points[:, 2], c='g', marker='o')
ax3.set_title('Mapped Back to Cube')
ax3.set_xlim([-1.5, 1.5])
ax3.set_ylim([-1.5, 1.5])
ax3.set_zlim([-1.5, 1.5])

plt.show()
```
